## Inheritance

```python
class person:
    def __init__(self,name,age):
        self.name= name 
        self.age = age 

    def introduce (self):
        print(f"My name is {self.name} and I am {self.age} years old") 
```

```python
p1 = person('Rutuja' , 23)
```

```python
class employee(person):

    def __init__(self,name,age,company):
        super().__init__(name,age)
        self.company = company

    def work(self):
        super().introduce()
        print(f"hi , I am working at {self.company}")
```

```python
emp2 = employee('sonu',35,'deloite')
```

```python
emp2.age
```

```python
emp2.name
```

```python
emp2.company
```

```python
emp2.introduce()
```

```python
emp2.work()
```

 ## Hierarchal inheritance

```python
# parent class 
class Bhagat_hyndai:

    def __init__(self,name,location):
        self.name = name 
        self.location = location

    def greeting(self):
        print(f"Hello{self.name} sir/madam .welcome to our showroom")

    def models_info(self):
        print(f"We have latest models available at our showroom like grand i10 and grand i20")
```

```python
# child class :child1
class grandi10(Bhagat_hyndai):

    def __init__(self,name,location,price_range):
        super().__init__(name,location)
        self.price_range = price_range

    def enquiry_price(self):
        super().greeting()
        print(f"Features of this model are .....and on-road price is 10lakhs")
        if self.price_range>10:
            print(f"Please check our best models falling under the same range:grand i20")

    def enquiry_models(self):
        super().models_info()
```

```python
# child class: child2
class grandi20(Bhagat_hyndai):

    def __init__(self, name, location,price_range):
        super().__init__(name, location)
        self.price_range = price_range
    
    def enquiry_price(self):
        super().greeting()
        print(f"Features of this model are ...... and on-road price is 15Lakhs")
    
    def enquiry_models(self):
        super().models_info()
```

```python
gi10_1 = grandi10('Pankaj','Ravet',12)
```

```python
gi10_1.price_range
```

```python
gi10_1.name
```

```python
gi10_1.location
```

```python
gi10_1.enquiry_price()
```

```python
gi10_2 = grandi10('naman','rahatani',8)
```

```python
gi10_2.enquiry_price()
```

```python
gi10_1.enquiry_models()
```

```python
gi20_1 = grandi20('chaman','kalewadi',15)
```

```python
gi20_1.name
```

```python
# parent class methods
gi20_1.greeting()
```

```python
# parent class methods
gi20_1.models_info()
```

```python
# methods of respective class 
gi20_1.enquiry_price()
```

```python
# methods of respective class 
gi20_1.enquiry_models()
```

Write a NumPy program to convert a list of numeric values into a one-dimensional NumPy array.

```python
import numpy as np 

l = [12.3,13.4,14.5,15.6,16.7]
print("Original List:" , l)

a = np.array(l)
print("one-dimensional numpy array:",a)
```
